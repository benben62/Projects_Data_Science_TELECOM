## 1 Régularisation de Tikhonov
On veut résoudre le problème suivant
$$\min_{w_o\in \mathbb{R},w\in \mathbb{R^p}}\frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i(x_i^Tw+w_0)))+\frac{\rho}{2}\|w\|_2^2$$

```python
def load_cervical_cancer(filename):
    """
    Cette fonction lit le fichier filename, par exemple
    filename = 'riskfactorscervicalcancer.csv'
    Elle retourne 
    X : une matrice de caracteristiques
    y : un vecteur des classes tel que si y[i] = 1, il y a un fort risque de cancer du col de l'uterus, et si y[i] = -1, les tests sont negatifs

    Pour plus d'infos sur la base de donnees,
    http://archive.ics.uci.edu/ml/datasets/Cervical+cancer+%28Risk+Factors%29#
    """

    data = np.loadtxt(filename, delimiter=',', skiprows=1)

    # predire si un des tests est positif
    y = np.max(data[:, -4:], axis=1) * 2 - 1
    # reequilibrer la base de donnees 
    pos_indices = np.where(y > 0)[0]
    neg_indices = np.where(y < 0)[0]
    indices = pos_indices.tolist() + neg_indices[1::7].tolist()

    y = y[indices]
    X = data[indices, :-4]

    # Standardisation de la matrice
    X = X - np.mean(X, axis=0)
    std_ = np.std(X, axis=0)
    X = X[:, std_ > 0]
    std_ = std_[std_ > 0]
    X = X / std_

    return X, y
```

### Question 1.1
We define that:<br><br>
$\tilde{x}_i = 
\begin{pmatrix}
1\\ 
x_i
\end{pmatrix}$

$\tilde{w} = 
\begin{pmatrix}
w_0\\ 
w
\end{pmatrix}$

$A=
\begin{pmatrix}
0 & \cdots  &\cdots   & 0\\ 
\vdots  &1  &  & \vdots  \\
\vdots&  &\ddots   & 0\\ 
0& \cdots  & 0  & 1
\end{pmatrix}$
<br><br>
So we can transform $f_1$ as:<br><br>
$$f_1:(\tilde{w}\in \mathbb{R}^{p+1})\mapsto \frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i\tilde{x}_i^T\tilde{w}))+\frac{\rho}{2}\|A\tilde{w}\|_2^2$$

And the gradient is:
$$
\begin{align}
\triangledown f_1 &= \frac{1}{n}\sum_{i=1}^n \frac{-y_i \cdot exp(-y_i\tilde{x}_i^T\tilde{w})}{1+exp(-y_i\tilde{x}_i^T\tilde{w})}\tilde{x}_i+\rho A \tilde{w}\\
&= \frac{1}{n}\sum_{i=1}^n (\frac{y_i \tilde{x}_i}{1+exp(-y_i\tilde{x}_i^T\tilde{w})}-y_i\tilde{x}_i)+\rho A \tilde{w}
\end{align}$$

And the Hessian matrix is:
$$\triangledown^2 f_1 = \frac{1}{n}\sum_{i=1}^n \frac{exp(-y_i\tilde{x}_i^T\tilde{w})}{(1+exp(-y_i\tilde{x}_i^T\tilde{w}))^2}(y_i^2\tilde{x}_i\tilde{x}_i^T))+\rho A$$
<br><br>
A is obviously positive-definite. For the left part, we can get that:<br><br>
$$\alpha^T \tilde{x}_i\tilde{x}_i^T \alpha = \|\tilde{x}_i^T \alpha\|^2_2 \geq 0$$
<br>
Where $\alpha$ is any vector of $\mathbb{R}^{p+1}$.<br><br>
What's more, $\frac{exp(-y_i\tilde{x}_i^T\tilde{w})}{(1+exp(-y_i\tilde{x}_i^T\tilde{w}))^2}y_i^2$ is obviously positive and $\rho A$ is also obviously positive-definite.<br><br>
So the Hessian matrix $\triangledown^2 f_1$ is positive-definite, which means the function is convex.
### Question 1.2

```python
X, y = load_cervical_cancer('riskfactorscervicalcancer.csv')
print X.shape, y.shape
```

```python
X1 = np.concatenate((np.ones((210, 1)), X), axis=1)
print X1.shape
```

```python
p = X1.shape[1]
n = X1.shape[0]
A = np.zeros((p, p), int)
np.fill_diagonal(A, 1)
A[0][0] = 0
def f1(w, rho, X, y):
    result = np.mean(np.log(np.exp(-y * X.dot(w)) + 1)) + rho/2.0*np.sum(A.dot(w) ** 2)
    return result
def gf1(w, rho, X, y):
    tmp = np.exp(-y * X.dot(w))
    result = - X.T.dot(tmp / (tmp + 1) * y)/n + rho * A.dot(w)
    return result
def ggf1(w, rho, X, y):
    result = np.zeros((p, p))
    for i in range(n):
        tmp = np.exp(-y[i] * X[i, :].dot(w))
        result = result + tmp / ((1 + tmp) ** 2) * (y[i] ** 2) * X[i, :].reshape((p, 1)).dot(X[i, :].reshape((1, p)))
    result = result / n + rho * A
    return result
```

```python
w = np.zeros(27)
print check_grad(f1, gf1, w, 1.0/210, X1, y)
```

We can see that the result of check_grad is very almost equal to 0, so the code we have written is right.

### Question 1.3
$$ w_{n+1} = w_{n} - (\triangledown ^2 f(w_n))^ {-1} \triangledown f(w_n)$$

```python
def newton(w0, rho, X, y):
    values = []
    w = w0
    f = gf1(w, rho, X, y)
    while np.sum(f ** 2) >= 1e-20:
        values.append(np.exp(np.sum(f ** 2)))
        g = ggf1(w, rho, X, y)
        w = w - np.linalg.inv(g).dot(f)
        f = gf1(w, rho, X, y)
    return w, values
```

```python
w0 = np.zeros(27)
w_newton, values_newton = newton(w0, 1.0/n, X1, y)
plt.figure()
plt.plot(values_newton)
plt.title('The norm of gradient as the iteration times increase')
plt.xlabel('The iteration times')
plt.ylabel('The norm of gradient')
plt.show()
print w_newton
print 'The minimun value of the method of Newton is: %.4f' % f1(w_newton, 1.0/n, X1, y)
```

```python
w1 = 0.3 * np.ones(27)
w_newton2, values_newton2 = newton(w1, 1.0/n, X1, y)
plt.figure()
plt.plot(values_newton2)
plt.title('The norm of gradient as the iteration times increase')
plt.xlabel('The iteration times')
plt.ylabel('The norm of gradient')
plt.show()
print 'The minimun value of the method of Newton is: %.4f' % f1(w_newton2, 1.0/n, X1, y)
```

### Response
We can observe that we cannot get a minimal value because it is out of region of quadratic convergence, so the Newton's method diverges.
### Question 1.5

```python
def lSNewton(w0, rho, X, y):
    values = []
    w = w0
    f = gf1(w, rho, X, y)
    t = np.sum(f ** 2)
    a = 0.5
    b = 1
    while t ** 0.5 >= 1e-10:
        values.append(np.exp(t))
        g = ggf1(w, rho, X, y)
        l = 0
        tmp = b * a ** l * np.linalg.inv(g).dot(f)
        while f1(w - tmp, rho, X, y) > f1(w, rho, X, y) - 0.5 * np.sum(tmp * f):
            l = l + 1
            tmp = b * a ** l * np.linalg.inv(g).dot(f)
        w = w - b * a ** l * np.linalg.inv(g).dot(f)
        f = gf1(w, rho, X, y)
        t = np.sum(f ** 2)
    return w, values
```

```python
w_newton3, values_newton3 = lSNewton(w1, 1./n, X1, y)
plt.figure()
plt.plot(values_newton3)
plt.title('The norm of gradient as the iteration times increase')
plt.xlabel('The iteration times')
plt.ylabel('The norm of gradient')
plt.show()
print w_newton3
print 'The difference of two results of question 1.3 and 1.5 is: %r' %(np.sum((w_newton3)**2)- np.sum((w_newton)**2))
print 'The minimum value of the method of Newton is: %.4f' % f1(w_newton3, 1.0/n, X1, y)
```

According to the result, we can see that $w^*$  and the minimum value that we got in this question is the same with those we got the question 1.3. And we used the similar iteration times to do the minimization. <br>
For the parameters, I have used the recommended parameters in the book and we cannot let b too small, otherwise it will take too many iteration times.

## 2 Régularisation pour la parcimonie
On s'intéresse toujours au problème de régression logistique mais on change la régularisation.
$$\min_{w_o\in \mathbb{R},w\in \mathbb{R^p}}\frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i(x_i^Tw+w_0)))+\rho \|w\|_1$$
### Question 2.1
We define that:<br><br>
$\tilde{x}_i = 
\begin{pmatrix}
1\\ 
x_i
\end{pmatrix}$

$\tilde{w} = 
\begin{pmatrix}
w_0\\ 
w
\end{pmatrix}$

$A=
\begin{pmatrix}
0 & \cdots  &\cdots   & 0\\ 
\vdots  &1  &  & \vdots  \\
\vdots&  &\ddots   & 0\\ 
0& \cdots  & 0  & 1
\end{pmatrix}$
<br><br>
So we can transform $f_2$ as:<br><br>
$$f_2:(\tilde{w}\in \mathbb{R}^{p+1})\mapsto \frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i\tilde{x}_i^T\tilde{w}))+\rho\|A\tilde{w}\|_1$$
We can find that the objectif function is *non-differentiable*, so we cannot use the Newton method.
### Question 2.2
Firstly, we define that:
$F_2 = f_2 + g_2$<br><br>
$f_2 = \frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i\tilde{x}_i^T\tilde{w}))$<br><br>
$g_2 = \rho\|A\tilde{w}\|_1$<br><br>
So, we can get the proximal operator of $g_2$
\begin{align}
prox_{g_2}(x) &= arg\min_{\tilde{w} \in \mathbb{R^{p+1}}}g_2(\tilde{w}) + \frac{1}{2\gamma}\|\tilde{w} - x\|^2\\
&= arg\min_{\tilde{w} \in \mathbb{R^{p+1}}} \rho\|A\tilde{w}\|_1 + \frac{1}{2 \gamma}\|\tilde{w} - x\|^2\\
&= \left\{\begin{matrix}
 x_i-\gamma \rho A_{ii} & x_i>\gamma \rho A_{ii}\\ 
 x_i+\gamma \rho A_{ii} & x_i<-\gamma \rho A_{ii}\\ 
 0 & -\gamma \rho A_{ii}\leq x_i \leq \gamma \rho A_{ii}& i \in [1, p+1]
\end{matrix}\right.
\end{align}<br>
and the gradient of $f_2$ is $$\triangledown f_2 = \frac{1}{n}\sum_{i=1}^n \frac{-y_i \cdot exp(-y_i\tilde{x}_i^T\tilde{w})}{1+exp(-y_i\tilde{x}_i^T\tilde{w})}\tilde{x}_i$$<br>
The objetif function is convex but not strictly convex.
### Question 2.3

```python
def funcTmp(x, A, rho, gamma):
    if x > gamma * rho * A:
        return x - gamma * rho * A
    elif x < - gamma * rho * A:
        return x + gamma * rho * A
    else:
        return 0
proxm = np.vectorize(funcTmp)

def prox(x, A, rho, gamma):
    res = proxm(x, A, rho, gamma)
    res[0] = x[0]
    return res

def f2(w, rho, X, y):
    result = np.mean(np.log(np.exp(-y * X.dot(w)) + 1))
    return result

def gf2(w, rho, X, y):
    tmp = np.exp(-y * X.dot(w))
    result = - X.T.dot(tmp / (tmp + 1) * y)/n
    return result

def F2(w, rho, X, y):
    result = np.mean(np.log(np.exp(-y * X.dot(w)) + 1)) + rho * np.sum(np.abs(A.dot(w)))
    return result

def gradientProx(w0, rho, X, y):
    values = []
    w = w0
    g = gf2(w, rho, X, y)
    value = f2(w, rho, X, y) + rho * np.sum(np.abs(A.dot(w)))
    t = np.sum(g ** 2)
    diff = 1
    a = 0.5
    b = 10
    while diff > 1e-5:
        l = 0
        pro = prox(w - b * a ** l * g, np.diag(A), rho, b * a ** l)
        while f2(pro, rho, X, y) > f2(w, rho, X, y) + np.sum(g * (pro - w)) + np.sum((w - pro)**2)/(2 * b * a ** l):
            l = l + 1
            pro = prox(w - b * a ** l * g, np.diag(A), rho, b * a ** l)
            b = 2 * b * a ** l
        w = prox(w - b * a ** l * g, np.diag(A), rho, b * a ** l)
        g = gf2(w, rho, X, y)
        t = np.sum(g ** 2) 
        value_old = value
        value = F2(w, rho, X, y)
        values.append(value)
        diff = abs(value - value_old)
    return w, values
```

```python
rho2 = 0.1
w2 = 0.5 * np.ones(27)
w_prox, values_prox = gradientProx(w2, rho2, X1, y)
plt.figure()
plt.plot(values_prox)
plt.title('The value of function as the iteration times increase')
plt.xlabel('The iteration times')
plt.ylabel('The value of function')
plt.show()
print w_prox
print 'The minimun value of the proximal gradient method is: %.5f' % F2(w_prox, rho2, X1, y)
```

I have chosen the difference of $x_{k+1}, x_{k}$ as the stopping condition. When the difference is smaller than $10^{-10}$, we stop the iteration.

1. The first question is differentiable and strictly convex with a penalization of $\|w\|_2^2$, which is kind of *Ridge* penalization.<br><br>
2. The second question is non-differentiable and convex with a penalization of $\|w\|_1$, which is kind of *Lasso* penalization.. However, we can spilit the objectif function into two functions like $F = f + g$, where $f = \frac{1}{n}\sum_{i=1}^n \log(1+exp(-y_i\tilde{x}_i^T\tilde{w}))$(differentiable and strictly convex) and $g =  \rho\|A\tilde{w}\|_1$(easy to compute the proximal operator).<br>
