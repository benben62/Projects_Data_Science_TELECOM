## Math question:
Let us consider L binary independent classifiers where the probability of a correct prediction is p > 0.5. Then, the prediction of the aggregation of models follows a Binomial distribution with parameters p and L (Why ?).

### Answer:
The prediction of the aggregation of models is:
$$\sum_{k=\lfloor{L/2}\rfloor+1}^L{\dbinom{L}{k}p^k(1-p)^{L-k}}$$
Assuming the binary classifiers are independent, it is the same question with throwing coins independently.
We need to make sure that $k>=\lfloor{L/2}\rfloor+1$ results of the base estimators are correct. For each k, we can get $\dbinom{L}{k}$ cases. And we add them all. The result follows a binomial distribution with parameters *p* and *L*.

## 1)

```python
L = 10
P = 0.7
rv = binom(L, P)
pre = 0
for i in xrange(L/2+1, L+1):
    pre += rv.pmf(i)
print 'The probability of correct prediction for the aggregation of models is: %.4f' %pre
```

## 1)

```python
# create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 1 * (0.5 - rng.rand(16))

n_estimators = 10 # L in the text
tree_max_depth = 10
bagging_max_depth = 10
# define the regressor by bagging stumps
tree = DecisionTreeRegressor(max_depth=tree_max_depth)
tree.fit(X, y)
bagging = BaggingRegressor(DecisionTreeRegressor(max_depth=1), n_estimators=n_estimators)
bagging.fit(X, y)
baggingDeeper = BaggingRegressor(DecisionTreeRegressor(max_depth=10), n_estimators=n_estimators)
baggingDeeper.fit(X, y)
# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_tree = tree.predict(X_test)
y_bagging = bagging.predict(X_test)
y_baggingDeeper = baggingDeeper.predict(X_test)
# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(X, y, 'o', c="k", label="Train data")
# add plots for Bagging/Tree
plt.plot(X_test, np.sin(X_test).ravel(), label="True date")
plt.plot(X_test, y_tree, label="tree, depth=10")
plt.plot(X_test, y_bagging, label="bagging, depth=1")
plt.plot(X_test, y_baggingDeeper, label="bagging, depth=10")
plt.title("Decision Tree Regression")
plt.legend(loc=1, numpoints=1)
plt.show()
```

## Analysis:
Here is the graph of decision tree and bagging. According to the graph, we can see that with the existence of noise, the 'bagging' performs better than 'tree'(it is closer to the real curve).

```python
maxDepth = 15
maxNumEsti = 30
argsDepth = xrange(1, maxDepth)
argsNumEsti = xrange(1, maxNumEsti)
scoresTree = []
for i in argsDepth:
    reg = DecisionTreeRegressor(max_depth=i)
    temp = np.mean(cross_val_score(reg, X=X, y=y, cv=5))
    scoresTree.append(temp)
plt.figure()
plt.plot(argsDepth, scoresTree)
plt.title('The tree depth')
plt.xlabel('tree_max_depth')
plt.ylabel('score')
plt.show()

scoresBagging = []
for i in argsNumEsti:
    reg = BaggingRegressor(n_estimators=i)
    temp = np.mean(cross_val_score(reg, X=X, y=y, cv=5))
    scoresBagging.append(temp)
plt.figure()
plt.plot(argsNumEsti, scoresBagging)
plt.title('The number of estimators')
plt.xlabel('L')
plt.ylabel('score')
plt.show()


scores = np.zeros((len(argsDepth), len(argsNumEsti)))
for i in argsDepth:
    for j in argsNumEsti:
        tree = DecisionTreeRegressor(max_depth=i)
        reg = BaggingRegressor(base_estimator=tree, n_estimators=j)
        score = np.mean(cross_val_score(reg, X, y, cv=5))
        scores[i-1][j-1] = score

argsX, argsY = np.mgrid[0:maxDepth-1, 0:maxNumEsti-1]
plt.figure(figsize=(15, 12))
ax = plt.subplot(111, projection='3d')
ax.plot_surface(argsX, argsY, scores, rstride=2, cstride=1, cmap=plt.cm.coolwarm, alpha=0.8)
ax.set_title('The 3D plot of scores')
ax.set_xlabel('The max depth of trees')
ax.set_ylabel('Number of estimators')
ax.set_zlabel('score')
plt.show()
```

## Analysis:
According to the first two 2D graphs, we can see that:
1. With the increase of the max_depth of decision tree, the score increase stably.
2. With the increase of the number of estimators, the change of score is quite unstable.

According to the 3D graphe, we can get the same result.

## 3)

```python
# Settings
n_repeat = 50       # Number of iterations for computing expectations
n_train = 50        # Size of the training set
n_test = 1000       # Size of the test set
noise = 0.1         # Standard deviation of the noise
np.random.seed(0)

# Change this for exploring the bias-variance decomposition of other
# estimators. This should work well for estimators with high variance (e.g.,
# decision trees or KNN), but poorly for estimators with low variance (e.g.,
# linear models).
estimators = [("Tree", DecisionTreeRegressor()),
              ("Bagging(Tree)", BaggingRegressor(DecisionTreeRegressor()))]

n_estimators = len(estimators)


# Generate data
def f(x):
    x = x.ravel()

    return np.sin(x)


def generate(n_samples, noise=1, n_repeat=1):
    rng = np.random
    X = np.sort(5 * rng.rand(n_samples, 1), axis=0)
    X = np.sort(X)
    if n_repeat == 1:
        y = f(X)
        y[::5] += noise*(0.5 - rng.rand(n_samples/5))
    else:
        y = np.zeros((n_samples, n_repeat))
        for i in range(n_repeat):
            temp = f(X)
            temp[::5] += noise*(0.5 - rng.rand(n_samples/5))
            y[:, i] = temp

    X = X.reshape((n_samples, 1))

    return X, y

X_train = []
y_train = []

for i in range(n_repeat):
    X, y = generate(n_samples=n_train)
    X_train.append(X)
    y_train.append(y)

X_test, y_test = generate(n_samples=n_test, n_repeat=n_repeat)

# Loop over estimators to compare
plt.figure(figsize=(15, 12))
for n, (name, estimator) in enumerate(estimators):
    # Compute predictions
    y_predict = np.zeros((n_test, n_repeat))

    for i in range(n_repeat):
        estimator.fit(X_train[i], y_train[i])
        y_predict[:, i] = estimator.predict(X_test)

    # Bias^2 + Variance + Noise decomposition of the mean squared error
    y_error = np.zeros(n_test)

    for i in range(n_repeat):
        for j in range(n_repeat):
            y_error += (y_test[:, j] - y_predict[:, i]) ** 2

    y_error /= (n_repeat * n_repeat)

    y_noise = np.var(y_test, axis=1)
    y_bias = (f(X_test) - np.mean(y_predict, axis=1)) ** 2
    y_var = np.var(y_predict, axis=1)

    print("{0}: {1:.4f} (error) = {2:.4f} (bias^2) "
          " + {3:.4f} (var) + {4:.4f} (noise)".format(name,
                                                      np.mean(y_error),
                                                      np.mean(y_bias),
                                                      np.mean(y_var),
                                                      np.mean(y_noise)))

    # Plot figures
    plt.subplot(2, n_estimators, n + 1)
    plt.plot(X_test, f(X_test), "b", label="$f(x)$")
    plt.plot(X_train[0], y_train[0], ".b", label="LS ~ $y = f(x)+noise$")

    for i in range(n_repeat):
        if i == 0:
            plt.plot(X_test, y_predict[:, i], "r", label="$\^y(x)$")
        else:
            plt.plot(X_test, y_predict[:, i], "r", alpha=0.05)

    plt.plot(X_test, np.mean(y_predict, axis=1), "c",
             label="$\mathbb{E}_{LS} \^y(x)$")

    plt.title(name)
    plt.ylim(-1., 1.5)
    if n == 0:
        plt.legend(prop={"size": 11})

    plt.subplot(2, n_estimators, n_estimators + n + 1)
    plt.plot(X_test, y_bias, "b", label="$bias^2(x)$"),
    plt.plot(X_test, y_var, "g", label="$variance(x)$"),
    plt.ylim(0., 0.1)
    if n == 0:
        plt.legend(prop={"size": 11})

plt.show()
```

## Analysis:
We calculate all the values with n_repeat times and take the average.
According to the results and the graphs, 
1. the $bias^2=0.0004$, so the estimators computed with the decision trees are biased.
2. the variance of Decision Tree is 0.0182, and the variance of Bagging is 0.0101. Obviously, the estimators based on bagging reduce the variance.
