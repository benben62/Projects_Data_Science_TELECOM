```python
import pandas as pd
import matplotlib.pyplot as plt
filename = 'household_power_consumption'
zipfilename = filename + '.zip'
na_values = ['?', '']
fields = ['Date', 'Time', 'Global_active_power']
df = pd.read_csv(filename + '.txt', sep=';', nrows=200000,
                 na_values=na_values, usecols=fields)
print df
# print df.dtypes
#11
print df[pd.isnull(df).Global_active_power]
#12
df = df.dropna()
print df
#13
df['Date'] = pd.to_datetime(df['Date'])
print df
#14
ts = df.loc[(df.Date >= '20070101') & (df.Date <= '20070430')].groupby(df.Date).mean()
plt.figure()
ts.plot()
plt.show()
```
